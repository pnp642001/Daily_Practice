```python
# Q7.Write a program for sum of digits.the digits are 76543 and the output should be 25.

def sum_of_string(num):
  sum=0
  while num >0:
    sum+=num % 10
    num = num // 10
  return sum

print(sum_of_string(76543))
```

```python
# Q5. Create a function that takes in a tuple of integers and returns the sum of the integers. Test the
# function with a tuple of your choice.

def tuple_sum(x):
  sum=0
  for num in x:
    sum+=num
  return sum

print(tuple_sum((4,5,6)))
```

```python
# Q3. Write a function called word_count(text) that takes a string as input and returns a dictionary where
# each key is a word in the text and its value is the number of times that word appears in the text. For
# example, word_count("hello world hello") should return {"hello": 2, "world": 1}.

def word_count(text):
  ans={}
  lst=text.split()

  for word in lst:
    if word in ans.keys():
      ans[word]+=1
    else:
      ans[word]=1
  return ans

print(word_count("hello world hello"))
```
